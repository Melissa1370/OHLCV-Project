```python
import yfinance as yf
import pandas as pd

# Define the cryptocurrency pairs (in USD)
symbols = ['BTC-USD', 'ETH-USD', 'BNB-USD', 'SOL-USD', 'XRP-USD']

# Define the period (30 days) and interval (hourly)
period = '1mo'
interval = '1h'

# Fetch data for each symbol
data = {}
for symbol in symbols:
    print(f"Downloading data for {symbol}...")
    ticker = yf.Ticker(symbol)
    df = ticker.history(period=period, interval=interval)
    
    # Add a 'coin_symbol' column
    df['coin_symbol'] = symbol.split('-')[0]  # Extract the symbol (e.g., BTC)
    
    # Reset the index to make 'timestamp' a column
    df.reset_index(inplace=True)
    
    # Select and rename columns to match the SQL table structure
    df = df[['coin_symbol', 'Datetime', 'Open', 'High', 'Low', 'Close', 'Volume']]
    df.columns = ['coin_symbol', 'timestamp', 'open', 'high', 'low', 'close', 'volume']
    
    # Append to the data dictionary
    data[symbol] = df

# Combine all DataFrames into a single DataFrame
crypto_ohlcv = pd.concat(data.values(), ignore_index=True)

# Convert 'timestamp' to datetime (if not already)
crypto_ohlcv['timestamp'] = pd.to_datetime(crypto_ohlcv['timestamp'])

# Ensure numeric columns are of type DECIMAL (float in pandas)
numeric_columns = ['open', 'high', 'low', 'close', 'volume']
crypto_ohlcv[numeric_columns] = crypto_ohlcv[numeric_columns].astype(float)

# Display the transformed DataFrame
print(crypto_ohlcv.head())
```
